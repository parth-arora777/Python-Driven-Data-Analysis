Data Cleaning



```python
#it shows count of duplicated values in datasets
net.duplicated().sum()
```

```python
#categorized on the basis of type column
net.groupby("type").type.count()
```

```python
#categorized on the basis of country column
net.groupby("country").type.count()
```

```python
#count nos of movies and series made in india
net[net["country"]=="India"].count
```

```python
#its shows all the tv shows release after 2005 
net[(net["release_year"]>2005) & (net["type"]=="TV Show")]
```

```python
#categorized on the basis of genre column
net.groupby("genre").type.count()
```

```python
#return the movie name with duration more than 100 minutes
net[(net["type"]=="Movie") & (net["duration"]>=100)]
```
